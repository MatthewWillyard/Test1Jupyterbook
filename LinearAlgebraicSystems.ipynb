## Linear Algebraic Systems


1. Solutions of Linear Systems



```python
from numpy.random import seed, choice, shuffle
from sympy import Matrix, latex, solve, zeros
from sympy.abc import x, y, z
from IPython.display import display, Math
from ipywidgets import interact
import ipywidgets as widgets

def f(seednum,numsols):

  P=range(1,10)
  seed(seednum)
  while True:
    A = Matrix(choice(P, (3, 4)))
    if numsols==1:
      if A[:, :3].rank() == 3 and A.rank() == 3:
        break
    elif numsols==0:
      if A[:, :3].rank() == 2 and A.rank() == 3:
        break
    else:
      if A[:, :3].rank() == 2 and A.rank() == 2:
        break

  A, b = A[:, :3], A[:, 3]
  u = Matrix([[x], [y], [z]])
  display(Math(f'{latex(A)}{latex(u)}={latex(b)}'))
  print(solve(A*u - b, [x, y, z]))



interact(f,
         seednum=widgets.IntText(value=1234, description='Seed='),
         numsols=widgets.Dropdown(options=[('One',1),('None',0),('Infinitely Many',2)],value=1,description="Number of Solutions="),
         );

```

```python
from numpy.random import seed, choice, shuffle
from numpy import identity
from sympy import Matrix, latex, solve, zeros, eye, Eq, MatMul
from IPython.display import display, Math
from ipywidgets import interact
import ipywidgets as widgets


def f(seednum,n,i,j,c):
  P=range(1,10)
  seed(seednum)
  A = Matrix(choice(P, (n, n)))
  Eijc=eye(n)
  Eijc[i-1,:]=Eijc[i-1,:]+c*Eijc[j-1,:]
  display(Math(r'A=%s' % latex(A)))
  display(Math(r'E_{%s%s}(%s)=%s'%(str(i),str(j),str(c),latex(Eq(MatMul(Eijc,A), Eijc*A)))))
  display(Math(r'E_{%s%s}(%s)A=%s'%(str(i),str(j),str(c),latex(Eq(MatMul(Eijc,A), Eijc*A)))))

interact(f,
         seednum=widgets.IntText(value=1234, description='Seed='),
         n=widgets.IntText(value=3,description='n='),
         i=widgets.IntText(value=1,description='i='),
         j=widgets.IntText(value=2,description='j='),
         c=widgets.IntText(value=3,description='c='),
         );
```

```python
from numpy.random import seed, choice, shuffle
from numpy import identity
from sympy import Matrix, latex, solve, zeros, eye, Eq, MatMul
from IPython.display import display, Math
from ipywidgets import interact
import ipywidgets as widgets

A=Matrix([[1,1,1]
        ,[1,-1,2]
        ,[2,1,-1]])

display(Math(r'A=%s' % latex(A)))

E1=Matrix([[1,0,0]
        ,[0,1,0]
        ,[-2,0,1]])

E2=Matrix([[1,0,0]
        ,[-1,1,0]
        ,[0,0,1]])

E3=Matrix([[1,0,0]
        ,[0,-1/2,0]
        ,[0,0,1]])

E4=Matrix([[1,0,0]
        ,[0,1,0]
        ,[0,1,1]])

E5=Matrix([[1,0,0]
        ,[0,1,0]
        ,[0,0,-2/7]])

E6=Matrix([[1,0,-1]
        ,[0,1,0]
        ,[0,0,1]])

E7=Matrix([[1,0,0]
        ,[0,1,1/2]
        ,[0,0,1]])

E8=Matrix([[1,-1,0]
        ,[0,1,0]
        ,[0,0,1]])

display(Math(r'I=%s' % latex(E8*E7*E6*E5*E4*E3*E2*E1*A)))



display(Math(r'A^{-1}=%s' % latex(Eq(MatMul(E8,E7,E6,E5,E4,E3,E2,E1),E8*E7*E6*E5*E4*E3*E2*E1))))


display(Math(r'A^{-1}=%s\approx%s' %( latex(A.inv()),latex(A.inv().evalf()))))

display(Math(r'A=%s' % latex(Eq(MatMul(E1.inv(),E2.inv(),E3.inv(),E4.inv(),E5.inv(),E6.inv(),E7.inv(),E8.inv()),E1.inv()*E2.inv()*E3.inv()*E4.inv()*E5.inv()*E6.inv()*E7.inv()*E8.inv()))))
```
